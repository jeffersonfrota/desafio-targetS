```python
#Versão alternativa
# Importando a biblioteca necessária
import matplotlib.pyplot as plt

# Para exibir o gráfico diretamente no Jupyter Notebook
%matplotlib inline

# Faturamento por estado
faturamento = {
    'SP': 67836.43,
    'RJ': 36678.66,
    'MG': 29229.88,
    'ES': 27165.48,
    'Outros': 19849.53
}

# Cálculo do total de faturamento
faturamento_total = sum(faturamento.values())

# Calculando o percentual de cada estado
percentuais = [(valor / faturamento_total) * 100 for valor in faturamento.values()]

# Criando o gráfico de pizza
labels = faturamento.keys()  # Nomes dos estados
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0']  # Cores para as fatias

plt.pie(percentuais, labels=labels, autopct='%1.2f%%', startangle=90, colors=colors)
plt.axis('equal')  # Assegura que o gráfico seja um círculo
plt.title('Percentual de Faturamento por Estado')

# Exibindo o gráfico
plt.show()
```
